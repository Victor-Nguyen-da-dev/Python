```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
test = df.pivot_table(index=['job_title_short', 'job_country'], aggfunc='size')
test.head(10)
```

```python
salary_stats = df.pivot_table(values='salary_year_avg', index= 'company_name', aggfunc=['max', 'min', 'mean'])
salary_stats.head(5)
```

```python
df_copy = df.copy()
df_copy = df_copy[df['job_title_short']=='Data Scientist']
sum_salary = df_copy.pivot_table(values='salary_year_avg', index=['company_name', 'job_country'], aggfunc= 'mean')
sum_salary = sum_salary[sum_salary['salary_year_avg']>200000]
sum_salary.head(10)
```
