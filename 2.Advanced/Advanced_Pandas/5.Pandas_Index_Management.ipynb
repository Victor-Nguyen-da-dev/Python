```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_1 = df.copy()
df_1.set_index('job_posted_date', inplace=True)
df_1.head(5)
```

```python
df_1.reset_index(drop=True, inplace=True)
df_1.head(5)
```

```python
results = df_1.pivot_table(values='salary_year_avg', index='company_name', aggfunc='median')
results = results.sort_index(ascending=False)
results.head(5)
```

```python
df_2 = df.copy()
pivot_test = df_2.pivot_table(values='salary_year_avg',index='job_title_short', aggfunc='median')
pivot_test.reset_index(inplace=True)
pivot_test.set_index('salary_year_avg',inplace=True)
pivot_test.head(5)
```
