```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_1 = df.copy()
df_1['job_postings_monthly'] = df_1['job_posted_date'].dt.to_period('M')
df_us_jobs = df_1[df_1['job_country'] == 'United States']
df_monthly_postings = df_1.groupby('job_postings_monthly').size().reset_index(name='Posting Count')

merged_df = pd.merge(df_us_jobs, df_monthly_postings, on='job_postings_monthly', how='right')
merged_df.head()
```

```python
df2 = df.copy()
df_not_remote_jobs = df2[df2['job_work_from_home'] == False]
df_not_remote_salary = df_not_remote_jobs.groupby('job_title_short')['salary_year_avg'].mean().reset_index()
df_not_remote_salary.rename(columns={'salary_year_avg': 'salary_not_remote_avg'}, inplace=True)

merged_df2 = pd.merge(df_not_remote_jobs, df_not_remote_salary, on='job_title_short', how='right')
merged_df2.head()
```

```python

df3 = df.copy()
df_company_salary = df3.groupby('company_name')['salary_year_avg'].mean().reset_index()
df_company_remote = df3[df3['job_work_from_home'] == True].groupby('company_name').size().reset_index(name='remote_count')

merged_df3 = pd.merge(df_company_salary, df_company_remote, on='company_name', how='inner')
merged_df3.head()
```
